```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np 

print(matplotlib.__version__)
print(np.__version__)
```

# Sample Data

```python
nu, sigma = 100, 15

x = nu + sigma * np.random.randn(10000)
x
```

```python
x.shape
```

# Histogram dengan pyplot Style

```python
plt.hist(x,
         bins=50,
         facecolor='g',
         alpha=0.75)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogran')

plt.text(45, 500, r'$\mu=100,\ \ sigma=15$')
plt.grid()

plt.show()
```

# Histogram dengan OO Style

```python
fig, ax = plt.subplots()

ax.hist(x,
         bins=50,
         facecolor='b',
         alpha=0.25)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogran')

plt.text(45, 500, r'$\mu=100,\ \ sigma=15$')
plt.grid()

plt.show()
```
